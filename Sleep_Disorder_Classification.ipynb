```python
path = "/content/drive/MyDrive/Colab Notebooks/Dataset/Sleep_health_and_lifestyle_dataset_with_nulls.csv"
dataset = pd.read_csv(path)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

# The imputation strategy:  mean, median, most_frequent, constant

impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

list_of_features = ["Heart Rate", "BMI Category"]

dataset[list_of_features] = impute.fit_transform(dataset[list_of_features])
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/Dataset/Sleep_health_and_lifestyle_dataset_preprocessed.csv"
dataset = pd.read_csv(path)
```

```python
dataset[['Systolic', 'Diastolic']] = dataset['Blood Pressure'].str.split('/', expand=True)

# Convert the new columns to numeric data types
dataset['Systolic'] = pd.to_numeric(dataset['Systolic'])
dataset['Diastolic'] = pd.to_numeric(dataset['Diastolic'])
```

```python
dataset.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

# Set up the LabelEncoder object
enc = LabelEncoder()

# Apply the encoding to the "Accessible" column
dataset['Sleep Disorder Encoder'] = enc.fit_transform(dataset['Sleep Disorder'])
dataset['Gender Encoder'] = enc.fit_transform(dataset['Gender'])
dataset['Occupation Encoder'] = enc.fit_transform(dataset['Occupation'])
dataset['BMI Category Encoder'] = enc.fit_transform(dataset['BMI Category'])
```

```python
dataset['Sleep Disorder Encoder'] = dataset['Sleep Disorder'].map({'None':0,
                                                                   'Sleep Apnea': 1,
                                                                   'Insomnia': 2})
dataset['Gender Encoder'] = dataset['Gender'].map({'Male':0,
                                                   'Female': 1})

dataset['Occupation Encoder'] = dataset['Occupation'].map({'Software Engineer':0,
                                                           'Doctor': 1,
                                                           'Sales Representative': 2,
                                                           'Teacher': 3,
                                                           'Nurse': 4,
                                                           'Engineer': 5,
                                                           'Accountant': 6,
                                                           'Scientist': 7, 'Lawyer': 8, 'Salesperson': 9})
dataset['BMI Category Encoder'] = dataset['BMI Category'].map({'Normal':0,'Normal Weight': 1, 'Overweight': 2, 'Obese': 3})


```

```python
dataset[["Sleep Disorder", "Sleep Disorder Encoder"]]
```

```python
dataset = dataset.drop(["Sleep Disorder"], axis=1)
dataset = dataset.drop(["Gender"], axis=1)
dataset = dataset.drop(["Occupation"], axis=1)
dataset = dataset.drop(["BMI Category"], axis=1)
dataset = dataset.drop(["Blood Pressure"], axis=1)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

list_of_features = ["Age", "Sleep Duration", "Quality of Sleep", "Physical Activity Level", "Stress Level", "Heart Rate", "Daily Steps", "Systolic", "Diastolic"]
scaler = MinMaxScaler()

dataset[list_of_features] = scaler.fit_transform(dataset[list_of_features])
```

# **Dataset Splitting**

- Training Set

- Testing Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x_data, y_data,
                                                    test_size = 0.2,
                                                    random_state = 1,
                                                    stratify = y_data)
```

# **Dataset Training**

Classification Problem

- Decision Tree

- K-nearest Neighbors

- Logistic Regression


## *K-nearest Neighbor*

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test) #testing
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score
knn_ac = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {round(accuracy_score(Y_test, Y_pred), 2)*100}%')
```

```python
knn_y_pred=knn.predict(X_test)
```

## *Logistic Regression*

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs', max_iter=2000)
```

```python
lr.fit(X_train, Y_train.values.ravel())
```

```python
Y_pred = lr.predict(X_test)
```

```python
Y_pred
```

```python
lr_ac = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {round(accuracy_score(Y_test, Y_pred), 2)*100}%')
```

```python
lr_y_pred=lr.predict(X_test)
```

## *Decision Tree*

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, Y_train) #training our model
```

```python
Y_pred = dt.predict(X_test) #testing our model
```

```python
Y_pred
```

```python
dt_ac = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {round(accuracy_score(Y_test, Y_pred), 2)*100}%')
```

```python
dt_y_pred=dt.predict(X_test)
```

# **Result analysis of used models**

```python
plt.bar(x = ['DT','KNN','LR'] , height = [dt_ac,knn_ac,lr_ac],color=['red', 'green', 'blue'])

plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
models_predictions=[knn_y_pred,lr_y_pred,dt_y_pred]
model={1:'KNN',2:'Logistic Regression',3:'Decision Tree'}


plt.figure(figsize=(15,7))
for i,y_pred in enumerate(models_predictions,1) :

    cm = confusion_matrix(y_pred,Y_test)

    plt.subplot(2,3,i)
    sns.heatmap(cm,cmap='BuPu',linewidth=3,fmt='',annot=True,
                xticklabels=['(None)','(Sleep_Apnea)','(Insomnia)'],
                yticklabels=['(None)','(Sleep_Apnea)','(Insomnia)'])


    plt.title(' CM of  '+ model[i])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.subplots_adjust(hspace=1,wspace=1)
```
